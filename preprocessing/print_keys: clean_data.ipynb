```python

from tqdm import tqdm

def clean_data(data):
    # Create a list to store the cleaned data
    cleaned_data = []
    
    # Loop through the items in the data
    for item in tqdm(data['data']):
        story = " ".join(item['story']).strip()  # Joining the story parts
        for question in item['questions']:
            q_id = question['q_id']
            q_type = question['q_type'].strip()
            q_qury = " ".join(question['query']).strip()
            question_text = question['question'].strip()
            answer = " ".join(question['answer']).strip()
            candidate_answers = " ".join(question['candidate_answers']).strip()
            question_info = question['question_info']
            reasoning_steps = question_info['reasoning_steps']
            num_facts=  question_info["num_facts"]
            asked_relation = " ".join(question_info['asked_relation']).strip() if isinstance(question_info['asked_relation'], list) else question_info['asked_relation']
            target_relation = " ".join(question_info['target_relation']).strip() if isinstance(question_info['target_relation'], list) else question_info['target_relation']
            
            cleaned_data.append({
                "q_id": q_id,
                "q_type": q_type,
                "q_query": q_qury,
                "story": story,
                "question": question_text,
                "answer": answer,
                "candidate_answers": candidate_answers,
                "reasoning_steps": reasoning_steps,
                "num_facts": num_facts,
                "asked_relation": asked_relation,
                "target_relation": target_relation
            })
    
    return cleaned_data

df= pd.DataFrame(clean_data(data))
df.to_csv("test_simple.csv", index=False)
```
